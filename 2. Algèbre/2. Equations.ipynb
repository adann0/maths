# Equations

- Inconnue : C'est une lettre qui désigne un nombre que l'on ne connait pas (par exemple : $x$).
- Equation : C'est une égalité qui contient une inconnue.
- Résoudre une Equation : C'est chercher et trouver la valeur de l'inconnue.
- Solution : C'est la valeur de l'inconnue.

\begin{equation}
a + b - b = a \\
\frac{a}{b} \cdot b = a
\end{equation}

## Exemples

**Exemple 1** :

- $2x + 10 = 30$
- $x = 10$

**Exemple 2** :

Prenons une equation $2x - 3 = 5$, résoudre une équation c'est trouver $x$, et donc d'avoir $x = nombre$.

Dans cette equation, on peut commencer par se débarasser du $3$ :

- $2x - 3 = 5$
- $2x = 5 + 3$
- $2x = 8$

A partir de la, on peut utiliser une **division** pour obtenir le résultat de l'équation :

- $2x/2 = 8/2$
- $x = 4$

**Exemple 3** : _Cas de l'équation-produit_

Une équation-produit est toujours égale à $0$.

Ce type d'équations à une propriété interessante :

- Si un produit de facteurs est nul, alors l'un au moins des facteurs est nul (Si $a*b=0$ alors $a=0$ ou $b=0$). Ce qui est logique vu qu'il n'y a que deux termes, et que l'on multiplie par 0, ca donne donc forcement 0.

Prenons l'équation $(x-5)(2x+3)=0$.

Il faut considérer que $a=(x-5)$ et que $b=(2x+3)$. On considère une fois que $a=0$ et une fois que $b=0$ en resolvant séparement les deux equations.

Pour la première :

- $x - 5 = 0$
- $x = 5$

Pour la seconde :

- $2x+3 = 0$
- $2x = -3$
- $x = -3/2$

On a donc ici les deux solutions de l'équation (oui il y en a deux).

**Exemple 4** : _Cas d'une equation contenant une fraction_

Prenons l'équation $\frac{x+4}{3} - \frac{x-1}{12} = \frac{3}{4}$.

Il faut dans un premier temps mettre toutes les fractions au même dénominateur (c'est a dire que toutes les divisions doivent avoir le même chiffre en bas). Supposons que l'on souhaite mettre tout les dénominateurs à 12 car on remarque que tous sont des multiples de 12.

Prenons le premier terme de l'équation : $\frac{x+4}{3}$

Il va falloir ici multiplier 3 par 4 pour obtenir 12 ; pour respecter les règles il va donc falloir simplement multiplier également le numérateur par le même nombre. Ce qui nous donne : $\frac{4*(x+4)}{4*3}$.

Prenons le dernier terme de l'équation : $\frac{3}{4}$.

Il va falloir ici multiplier 4 par 3 (et pas par 4) pour obtenir 12 ; il va donc falloir multiplier le numérateur également par 3 (et pas par 4) ; ce qui nous donnera : $\frac{3*3}{3*4}$.

- $\frac{x+4}{3} - \frac{x-1}{12} = \frac{3}{4}$
- $\frac{4*(x+4)}{4*3} - \frac{x-1}{12} = \frac{3*3}{3*4}$
- $\frac{4*(x+4)}{12} - \frac{x-1}{12} = \frac{9}{12}$

On peut maintenant enlever toutes les fractions :

- $4*(x+4) - (x-1) = 9$
- $4x + 16 - x + 1 = 9$
- $3x + 17 = 9$
- $3x = 9 - 17$
- $3x = -8$
- $3x/3 = -8/3$
- $x = -8/3$

## A quoi ça sert ?

Un exemple simple est la resolution d'un problème. Supposons que l'on souhaite connaitre le temps de téléchargement d'un fichier de 2Go alors que l'on a une vitesse de connexion de 5mo/s.

Si $x$ est le temps d'attente :

$2000 = 5x$

$2000/5 = x$

$x = 400$

Il faudra attendre environs 400 secondes pour que le téléchargement soit effectué. Télécharger _5_ mo par secondes durant _400_ secondes nous donneraient bien _2000_ mo donc on a trouvé la solution de l'équation représentant ce problème.
