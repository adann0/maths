# Oprations sur les fractions

Dans le cas d'une simple opration sur une fraction, on ne peut pas supprimer le dnominateur. 

On peut supprimer le dnominateur uniquement lorsque on a une equation, et que l'on a mis les deux cots de l'quation au mme denominateur. Etant donn que l'on a "multipli le cot droit" pour qu'il atteigne le mme dnominateur que le cot gauche, l'quation reste juste et proportionnelle, alors que supprimer les denominateurs d'un calcul normal modifierais juste le rsultat sans conserver la proportionalit (maintenue justement par les fractions). 

## Addition et soustraction



Lorsqu'on additionne deux fractions qui ont le **mme dnominateur**, on additionne les numrateurs _a_ + _b_ et on garde le dnominateur _D_ :

\begin{equation}
\frac{a}{D} + \frac{b}{D} = \frac{a+b}{D}
\end{equation}

Lorsqu'on soustraie deux fractions qui ont le **mme dnominateur**, on additionne les numrateurs _a_ - _b_ et on garde le dnominateur _D_ :

\begin{equation}
\frac{a}{D} - \frac{b}{D} = \frac{a-b}{D}
\end{equation}

**Exemple** :

_Sinon on souhaite additionner ou soustraire deux fractions qui n'ont pas le mme dnominateur, il faut dans un premier temps les mettre au mme dnominateur en trouvant un multiple commun entre les deux dnominateurs :_

- $A = \frac{4}{3} + \frac{2}{9}$
- $\ \ \ = \frac{4\times3}{3\times3} + \frac{2}{9}$
- $\ \ \ = \frac{12}{9} + \frac{2}{9}$
- $\ \ \ = \frac{14}{9}$

- $B = \frac{1}{3} - \frac{5}{4}$
- $\ \ \ = \frac{1 \times 4}{3 \times 4} - \frac{5 \times 3}{4 \times 3}$
- $\ \ \= \frac{4}{12} - \frac{15}{12}$
- $\ \ \= \frac{-11}{12}$

## Multiplication

Lorsqu'on multiplie des fractions, on multiplieles numrateurs entre et et les dnominateurs entre eux :

\begin{equation}
\frac{a}{b} + \frac{c}{d} = \frac{a\times c}{b\times d}
\end{equation}

**Exemple** :

- $C = \frac{7}{18} \times \frac{81}{56}$
- $\ \ \ = \frac{7 \times 81}{18 \times 56}$

_A ce niveau, avant de multiplier les dnominateurs et les numrateurs entre eux, il est bon de simplifier en trouvant des multiples (a peut se faire en diagonale ou de haut en bas) :_

- $\ \ \ = \frac{7 \times 9 \times 9}{9 \times 2 \times 7 \times 8}$
- $\ \ \ = \frac{9}{2 \times 8}$
- $\ \ \ = \frac{9}{16}$

## Division

| L'inverse de ...  |Est ...               | $$\times$$           |
|:-----------------:|:---------------------:|:---------------------|
| $$3$$             | $$\frac{1}{3}$$       | $$1$$
| $$\frac{1}{2}$$   | $$\frac{2}{1} = 2$$   | $$1$$
| $$\frac{7}{12}$$  | $$\frac{12}{7}$$      | $$1$$
| $$x$$             | $$\frac{1}{x}$$       | $$1$$

Deux nombres sont **inverses** l'un de l'autre si leur produit est gal  1 :

\begin{equation}
\frac{a}{b} \cdot \frac{b}{a} = 1 \\
\end{equation}

Diviser par un nombre c'est multiplier par son inverse :

\begin{equation}
\frac{a}{b} \div \frac{c}{d} = \frac{a}{b} \times \frac{d}{c}
\end{equation}

**Exemple** :

- $D = 4 \div 8 = 0.5$
- $\ \ \ = 4 \times \frac{1}{8} = 0.5$

- $E = \frac{2}{3} \div \frac{5}{4}$
- $\ \ \= \frac{2}{3} \times \frac{4}{5}$
- $\ \\= \frac{8}{15}$
