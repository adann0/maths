## Dfinitions

### Graphe



_(Un graphe avec trois sommets et trois artes)._

Un **graphe simple non orient** est un couple _G_ = (_V_, _E_) comprenant :

- _V_ un ensemble de sommets (aussi appels nuds, points ou vertex) ;
- _E_  {{_x_, _y_} | (_x_, _y_)  V  x  y} un ensemble d'artes (aussi appeles liens ou lignes), qui sont des paires de sommets (c.--d. qu'une arte est associe  deux sommets distincts).

Dans un graphe simple non orient d'ordre n, le degr maximum d'un sommet est _n  1_ et la taille maximale du graphe est _n(n  1)/2_. 

Dans l'arte {x, y}, les sommets x et y sont appels les extrmits ou les sommets extrmes de l'arte.

- _V_ un ensemble de sommets (aussi appels nuds, points ou vertex) ;

### Graphe orient



_(Un graphe orient avec trois sommets et quatre artes)._

Un **graphe simple orient** est un graphe dans lequel les artes possdent une orientation. Un graphe orient est un couple G = (V, A) (parfois G = (V, E)) comprenant :

- _V_ un ensemble de sommets (aussi appels nuds ou points) ;
- _A_  {(_x_, _y_) | (_x_, _y_)  V  x  y} un ensemble de flches (aussi appeles artes orientes  parfois simplement artes avec l'ensemble correspondant nomm E au lieu de A , liens orients ou lignes orientes), qui sont des couples de sommets distincts (c.--d. une flche est associe  deux sommets distincts).

Dans la flche (x, y) oriente de x vers y, x est appel la queue de la flche et y la tte de la flche. La flche (y, x) est appele la flche inverse de (x, y).

Il existe trois grandes familles de graphes et cinq catgories au total :

## Vocabulaire

- **Complet** : un graph est complet si tout les sommets sont relis a tout les sommets.



- **Voisin** : on dit que les sommets $u$ et $v$ sont voisin si ils sont relis par une arrte (si ils ne sont _pas_ relis par une arrte, on dit qu'ils ne sont _pas voisins_).
- **Degr** : le degr d'un sommet est le nombre de liens (artes ou arcs) reliant ce sommet, avec les boucles comptes deux fois. Le degr d'un sommet $s$ est not $\deg(s)$. 
- **Degr maximal/Degr minimal** : Le degr maximal d'un graphe $G$, not $\Delta(G)$, et le degr minimal de ce graphe, not $\delta(G)$, sont respectivement le maximum et le minimum des degrs des sommets du graphe.



- **Chemin** : C'est une suite d'arrte qui permettent de relier deux sommets (il peut y avoir plusieurs chemins, de longueur differente).
- **Longueur d'un chemin** : Son nombre d'arrtes.



- **Cycle** : un chemin dont les 2 extrmits sont relies (sous entendu, par une boucle, on ne passe pas deux fois par la meme arrte).



- **Connexe** : pour tout $u$ et $v$, si le graphe contient un chemin entre $u$ et $v$ alors le graphe est connexe (un graphe est connexe si tout les sommets ont au moins un chemin vers tout les sommets). Le premier graphe ci-dessous est connexe, le second n'est pas connexe, et tout deux sont bien des graphes valides.





- **Arbre** : un arbre est un graphe connexe et sans cycle. Le premier graphe ci-dessous est un arbre, le second n'est pas un arbre.





## Algorithme de parcours en largeur

L'algorithme de parcours en largeur (ou BFS, pour Breadth-First Search en anglais) permet le parcours d'un graphe ou d'un arbre de la manire suivante : on commence par explorer un nud source, puis ses successeurs, puis les successeurs non explors des successeurs, etc. 

L'algorithme de parcours en largeur permet de calculer les distances de tous les nuds depuis un nud source dans un graphe non pondr (orient ou non orient). Il peut aussi servir  dterminer si un graphe non orient est connexe.



Un parcours en largeur dbute  partir d'un nud source. Puis il liste tous les voisins de la source, pour ensuite les explorer un par un. Ce mode de fonctionnement utilise donc une file dans laquelle il prend le premier sommet et place en dernier ses voisins non encore explors. Les nuds dj visits sont marqus afin d'viter qu'un mme nud soit explor plusieurs fois. Dans le cas particulier d'un arbre, le marquage n'est pas ncessaire. Voici les tapes de l'algorithme :

- mettre le nud source dans la file ;
- retirer le nud du dbut de la file pour le traiter ;
- mettre tous ses voisins non explors dans la file ( la fin) ;
- si la file n'est pas vide reprendre  l'tape 2.

L'algorithme de parcours en profondeur diffre du parcours en largeur car il continue l'exploration jusqu' arriver dans un cul-de-sac. C'est seulement  ce moment-l qu'il revient en arrire pour explorer depuis un autre nud voisin. L'utilisation d'une pile au lieu d'une file transforme l'algorithme du parcours en largeur en l'algorithme de parcours en profondeur.
