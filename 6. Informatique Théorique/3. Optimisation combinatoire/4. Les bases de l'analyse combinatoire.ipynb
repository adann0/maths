# Bases de l'analyse combinatoire

## Principe multiplicatif

**Théorème** : On considère deux ensembles finis $A$ et $B$. On note $A \times B$ le produit de A par B, c'est-à-dire l'ensemble des couples $(x, y)$ ou $x$ est dans $A$ et $y$ est dans $B$. Alors : $card(A \times B) = card(A) \times card(B)$.

**Exemple** :

Un magasin propose 3 téléphones de même dimension et 4 coques compatibles mais différentes.

De combien de choix possible dispose-t-on ?

- 3 * 4 = 12

On dispose de 12 choix possible.

## Arrangements (choix en tenant compte de l'ordre)

### Arrangements sans répétition

**Théorème** — Le nombre d'arrangements sans répétition de _n_ éléments pris _k_ à _k_ est égal à $A_{n}^{k}$ (égal à ${\frac {n!}{(n-k)!}}$ si _k_≤_n_ et à 0 sinon).

\begin{equation}
{\frac {n!}{(n-k)!}}
\end{equation}

**Exemple** :

On organise une course comprenant 10 personnes et on s'intérese au podium formé des 3 premiers dans l'ordre. Combien de podiums différents existe-t-il ?

- $n = 10$
- $k = 10$
- ${\frac {10!}{(10-7)!}} = 10 \times 9 \times 7 = 720$

Il existe 720 podiums différents.

### Arrangements avec répétition (= principe multiplicatif)

**Théorème** : Le nombre d'arrangements avec répétition de $k$ objets parmi $n$ objets est : 

\begin{equation}
n^k
\end{equation}

(C'est le cas ou les coureurs pourraient se cloner, ce qui dans ce cas serait illogique, mais dans d'autres cas essentiel, par exemple si on souhaite bruteforcer un mot de passe, il faudrait pouvoir faire "aaa" (donc arrangement avec répétition)).

**Exemple** :

Combien de nombres à 5 chiffres peut-on écrire avec les chiffres 1, 2 et 3 seulement ? 

## Permutations

**Théorème** : Le nombre de permutations de $n$ objets distincts est : 

\begin{equation}
n!
\end{equation}

(C'est le cas ou les coureurs seraient que 3, ou qu'il y a autant de place sur le podium que de coureurs).

**Exemple** :

Combien de permutations peut ont réaliser avec les objets "4", "5" et "8" ? 

- 458, 485, 584, 548, 845, 854
- $3! = 3 \times 2 \times 1 = 6$

## Combinaisons (choix sans tenir compte de l'ordre)

**Théorème** : Le nombre de combinaisons sans répétition de $k$ objets parmi $n$ objets est : 

\begin{equation}
C_{n}^{k}\,={\frac {n!}{k!(n-k)!}}
\end{equation}

(C'est le cas ou on ne prend pas en compte l'ordre : [1, 2, 3] == [3, 2, 1] == [2, 1, 3] == [2, 3, 1] == ... ; donc on se passe d'une partie des résultats qui pour une même "combinaison" seraient identiques). Ca serait extremement weak pour un bruteforcer parceque pour "abc" on aurait plus jamais les lettres "a" "b" et "c" vu que la combinaison aurait déjà été effectué).
