## fonctions basiques

### fonctions linéaires

#### propriétés

    f(x) = mx + b


Les fonctions linéaires sont des fonctions ou la varriable d'entrée de la fonction est x. Les lettres m et b sont des constantes qui sont données.

- m (pente de la fonction) : si la pente est positive, la fonction est croissante ; si la pente est négative, la fonction est décroissante. m est multiplié à x, donc x est augmenté autant de fois que la grandeur de la pente.

- b (ordonnée à l'origine) : si on met x égale a 0, la sortie de la fonction sera égale à b

Elles sont utilisées IRL dans la physique, l'économie et les sciences sociales. Par exemple, pour calculer la demande et l'offre de biens et de services ; ou la relation entre des variables comme le revenu et l'éducation, ou le temps et la satisfaction au travail.

#### trace

```python
import matplotlib.pyplot as plt
import numpy as np

m = 1
b = 0

# Définition des points pour la fonction linéaire
x = np.linspace(0, 10, 100) # Crée 100 points de 0 à 10
y = m * x + b # Fonction linéaire y = 1x + 0

# Création de la figure
fig, ax = plt.subplots()

# Ajout de la courbe
ax.plot(x, y)

# Ajout des lignes en abscisse et en ordonnée
ax.grid(True, which='both')

# Ajout des ticks
ax.tick_params(axis='x', which='both', direction='in', bottom=True, top=True)
ax.tick_params(axis='y', which='both', direction='in', left=True, right=True)

# Ajout des titres et étiquettes d'axe
ax.set_title('Fonction linéaire y = 1x + 0')
ax.set_xlabel('x')
ax.set_ylabel('y')

# Affichage de la figure
plt.show()
```
