# Probabilité

## Vocabulaire

- Equiprobabilité : En théorie des probabilités et en statistique, l'équiprobabilité de deux évènements signifie que ces deux évènements ont une même probabilité.
- Epreuve aléatoire : une epreuve aléatoire est une épreuve dont l'issue n'est pas connue à l'avance (par exemple : le lancé d'un dé).
- Univers (Ω) : l'univers est l'ensemble des résultats possibles d'une épreuve aléatoire (dans l'exemple du dé : Ω = {1, 2, 3, 4, 5, 6}).
- Evènement (A) : un evènement est une partie de l'univers. un évènement est souvent une proposition, en lien avec l'univers (c'est à dire avec le ou les résultats possibles) (dans l'exemple du dé, un évènement A pourrait être "obtenir un résultat pair", il serait écrit A = {2, 4, 6}). L'évènement contraire de A est noté Ā (dans l'exemple ci-dessus, Ā = {1, 3, 5}).
- Evènement certain : On réalise l'évènement certain lorsqu'on réalise Ω (ensemble oméga) (dans l'exemple du lancé de dé, si on prend un évènement C "obtenir un résultat inférieur à 8", il serait écrit C = {1, 2, 3, 4, 5, 6} = Ω).
- Evènement impossible : On réalise l'évènement impossible lorsqu'on réalise ∅ (ensemble vide) (dans l'exemple du lancé de dé, si on prend l'évènement I "obtenir un résultat supérieur à 7", il serait écrit I = {} = ∅).
- Cardinal : C'est la _len_ d'un ensemble (par exemple le Cardinal de C dans l'exemple précédent est 6).

## Opérations

On reprend l'exemple du lancé de dé.

On défini 3 évènements :

- A : "obtenir un nombre pair" (A = {2, 4, 6})
- B : "obtenir un multiple de 3" (B = {3, 6})
- C : "obtenir un diviseur de 3" (C = {1, 3})

## Union/Intersection

- **Union** : C'est le Ou Inclusif (ça serait "obtenir un nombre pair OU obtenir un multiple de 3" ; donc A ∪ B = {2, 3, 4, 6})
- **Intersection** : C'est le Et (ca serait "obtenir un nombre pair ET obtenir un multiple de 3" ; donc A ∩ B = {6})

## Cardinal

L'union et l'intersection de deux parties $A$ et $B$ de $E$ sont reliées par la formule :

- ${\mathrm {Card}}(A\cup B)={\mathrm {Card}}(A)+{\mathrm {Card}}(B)-{\mathrm {Card}}(A\cap B)$.

## Exemple

On a une classe de 30 élèves.

- 20 élèves étudient l'anglais.
- 7 élèves étudient l'espagnol.
- 3 élèves étudient les deux langues.

Combien d'élèves n'étudient aucune langue ? 

Spoiler : On sait que l'on a 3 élèves dans les deux langues. On sait que l'on a en tout 20 élèves qui étudient l'anglais, il y en a donc 17 qui n'étudient que l'anglais. 7 élèves étudient l'espagnol, donc il y en a 4 qui n'étudient que l'espagnol. 3 se trouvent dans les deux categories donc 6 élèves n'étudient aucune langue.

|:-----:|:-:|:-:|:-:|
| **A** | 3 | 17| 20|
| **Ā** | 4 | 6 | 10|
| **∑** | 7 | 23| 30|

## Probabilité

La probabilité d'un évènement A notée P(A) est un nombre compris entre 0 et 1 : 

\begin{equation}
0 \le P(A) \le 1
\end{equation}

et tel que :

- $P(Ω) = 1$
- $P(\emptyset )=0$
- $P(\overline A)=1-P(A)$
- $P(A\cup B)=P(A)+P(B)-P(A\cap B)$

## Equiprobabilité

Si le dé est équilibré c'est a dire qu'on se place en situation d'équiprobabilité (on a $P(\{1\}) = P(\{2\}) = ... = P(\{6\}) = \frac{1}{6}$).

La probabilité de l'événement A est alors donnée par la formule

\begin{equation}
P(A)={\dfrac {\operatorname {Card}(A)}{\operatorname {Card}(\Omega )}}={\dfrac {{\text{nombre de cas favorables}}}{{\text{nombre de cas possibles}}}}
\end{equation}
