# Suites Géométriques

Considérons une suite numérique $(u_n)$ ou le _rapport_ entre un terme et son précédent reste constant et égale à 2 dont le premier terme est égal à 5.

| $$n$$     | _0_   | _1_   | _2_   | _3_   | _n_   | $$u_{n+1}$$ |
|:---------:|:-----:|:-----:|:-----:|:-----:|:-----:|:-----------:|
| $$(u_n)$$ | 5     | 10    | 20    | 40    | $$5*(2^n)$$| $u_n * 2$ |

Une telle suite est appelée suite géométrique de raison 2 et de premier terme 5.

**Définition 1** :

Une suite $(u_n)$ est une **suite géométrique** s'il existe un nombre $q$ tel que pour tout entier $n$, on a : 

\begin{equation}
u_{n+1} = q * u_n 
\end{equation}

Le nombre $q$ est appelé **raison** de la suite.

**Définition 2** :

Si $(u_n)$ est une suite géométrique de raison $q$ et de premier terme $u_0$. Pour tout entier naturel $n$, on a : 

\begin{equation}
u_n = u_0 * q^n 
\end{equation}

**Définition 3** : _Variations d'une suite arithmétique_

$(u_n)$ est une suite géométrique de raison $q$ et de premier terme non nul $u_0$.

Pour $u_0 > 0$ :

- Si $q > 0$ alors la suite $(u_n)$ est croissante.
- Si $0 < q < 1$ alors la suite $(u_n)$ est décroissante.

Pour $u_0 < 0$ :

- Si $q > 0$ alors la suite $(u_n)$ est décroissante.
- Si $0 < q < 1$ alors la suite $(u_n)$ est croissante.

**Exemple** : _Variations d'une suite géométrique_





```python
import matplotlib.pyplot as plt

# Définition de la suite
def u(n):
    return 1+1.1**n

# Création de la liste des valeurs de la suite pour n allant de 0 à 10
n_values = list(range(11))
u_values = [u(n) for n in n_values]

# Tracé de la suite
plt.plot(n_values, u_values, 'bo--')

# Ajout de labels pour les axes
plt.xlabel('n')
plt.ylabel('u(n)')

plt.title("Suite géométrique croissante (r > 1) (u0 = 1)")


# Affichage de la figure
plt.show()
```

## Sommes de termes

\begin{equation}
1 + q + q^2 + q^3 + ... + q^n = \frac{1-q^{n+1}}{1-q}
\end{equation}
