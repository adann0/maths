# Fonctions

## Vocabulaire et notations

Pour bien comprendre, on va partir d'une situation concrète. On te dit que "Le prix d'entrée à un spectacle est de 20 euros par personne".

A partir de cette phrase, on peut déduire ce _tableau de valeurs_ :

| $x$ (Nombre de personnes) | 1 | 2 | 3 | 5 | 10 |
|-----------------------|---|---|---|---|----|
| $P$ (Prix)                 | 20| 40| 60| 100| 200|

On remarque que l'on a une correspondance entre x (le nombre de personne) et P (le prix a payer).

- $1 \mapsto 20$

- $2 \mapsto 40$

On peut déterminer la formule de la fonction (les 3 notations pour une fonction qui reviennent au même) :

- $P : x \mapsto 20x$

- $P(x) = 20x$

- $y = 20x$

Une fonction est donc une machine mathématique, qui prend un nombre, transforme ce nombre (via certaines règles, la formule - ici $20x$) et renvoi un nombre. $x$ est appelé une varriable car $x$ peut varrier.

### Fonction croissante/décroissante

- Dire que $f$ est **croissante** signifie que : si $a < b$ alors $f(a) \leq f(b)$.




```python
import matplotlib.pyplot as plt
import numpy as np

# Définition de la fonction linéaire
def lin(x):
    return x

# Génération des valeurs pour x et y
x = np.linspace(-5, 5, 100)
y = lin(x)

# Création du graphique
fig, ax = plt.subplots()

# Tracer la courbe
ax.plot(x, y)

# Centrer l'axe des ordonnées et des abscisses sur la courbe
ax.spines['left'].set_position('zero')
ax.spines['right'].set_color('none')
ax.yaxis.tick_left()

ax.spines['bottom'].set_position('zero')
ax.spines['top'].set_color('none')
ax.xaxis.tick_bottom()

# Afficher les points (a, f(a)) et (b, f(b)) pour a < b
a = -2
b = 2
fa = lin(a)
fb = lin(b)
ax.scatter(a, fa, color='r', label=f'({a}, {fa})')
ax.scatter(b, fb, color='g', label=f'({b}, {fb})')

# Ajouter une légende
ax.legend()

# Ajout des labels et titre
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Fonction affine (strictement croissante) (x)')

# Tracer les lignes pointillées

ax.plot([a,a], [0,fa], color='r', linestyle='--')
ax.plot([b,b], [0,fb], color='g', linestyle='--')
ax.plot([a,0], [fa,fa], color='r', linestyle='--')
ax.plot([b,0], [fb,fb], color='g', linestyle='--')

# Affichage du graphique
plt.show()
```

- Dire que $f$ est **décroissante** signifie que : si $a < b$ alors $f(a) \geq f(b)$.

```python
import matplotlib.pyplot as plt
import numpy as np

# Définition de la fonction linéaire
def lin(x):
    return -x

# Génération des valeurs pour x et y
x = np.linspace(-5, 5, 100)
y = lin(x)

# Création du graphique
fig, ax = plt.subplots()

# Tracer la courbe
ax.plot(x, y)

# Centrer l'axe des ordonnées et des abscisses sur la courbe
ax.spines['left'].set_position('zero')
ax.spines['right'].set_color('none')
ax.yaxis.tick_left()

ax.spines['bottom'].set_position('zero')
ax.spines['top'].set_color('none')
ax.xaxis.tick_bottom()

# Afficher les points (a, f(a)) et (b, f(b)) pour a < b
a = -2
b = 2
fa = lin(a)
fb = lin(b)
ax.scatter(a, fa, color='r', label=f'({a}, {fa})')
ax.scatter(b, fb, color='g', label=f'({b}, {fb})')

# Ajouter une légende
ax.legend()

# Ajout des labels et titre
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Fonction affine (strictement décroissante) (-x)')

# Tracer les lignes pointillées

ax.plot([a,a], [0,fa], color='r', linestyle='--')
ax.plot([b,b], [0,fb], color='g', linestyle='--')
ax.plot([a,0], [fa,fa], color='r', linestyle='--')
ax.plot([b,0], [fb,fb], color='g', linestyle='--')

# Affichage du graphique
plt.show()
```

- Dire que $f$ est **constante** signifie que : $f(a) = f(b)$.

```python
import matplotlib.pyplot as plt
import numpy as np

# Définition de la fonction linéaire
def lin(x):
    return 0*x+1

# Génération des valeurs pour x et y
x = np.linspace(-5, 5, 100)
y = lin(x)

# Création du graphique
fig, ax = plt.subplots()

# Tracer la courbe
ax.plot(x, y)

# Centrer l'axe des ordonnées et des abscisses sur la courbe
ax.spines['left'].set_position('zero')
ax.spines['right'].set_color('none')
ax.yaxis.tick_left()

ax.spines['bottom'].set_position('zero')
ax.spines['top'].set_color('none')
ax.xaxis.tick_bottom()

# Afficher les points (a, f(a)) et (b, f(b)) pour a < b
a = -2
b = 2
fa = lin(a)
fb = lin(b)
ax.scatter(a, fa, color='r', label=f'({a}, {fa})')
ax.scatter(b, fb, color='g', label=f'({b}, {fb})')

# Ajouter une légende
ax.legend()

# Ajout des labels et titre
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Fonction affine (fonction constante) (0x + 1)')

# Tracer les lignes pointillées

ax.plot([a,a], [0,fa], color='r', linestyle='--')
ax.plot([b,b], [0,fb], color='g', linestyle='--')
ax.plot([a,0], [fa,fa], color='r', linestyle='--')
ax.plot([b,0], [fb,fb], color='g', linestyle='--')

# Affichage du graphique
plt.show()
```

- Dire que $f$ est **monotone** signifie que $f$ est soit croissante, soit décroissante.

- On dit qu'une fonction croissante conserve l'ordre.
- On dit qu'une fonction décroissante renverse l'ordre.

### Extrémité/Maximum/Minimum/Interval

On a ici ci-dessous une fonction qui est d'abord décroissante et ensuite croissante.

Cela provoque un minimum, on peut lire la valeur de ce minimum, 0. On peut également dire pour quelle valeur de x il est atteint (0).

- Dire que $f$ admet un maximum _M_ en $a$ signifie que : pour tout $x$, $f(x) \leq M = f(a)$.
- Dire que $f$ admet un minimum _m_ en $b$ signifie que : pour tout $x$, $f(x) \geq m = f(b)$.

Ce qui signifie donc que pour la fonction carré, on aura toujours un résultat supérieur ou égal à 0.

Parfois c'est vrai sur un interval de la fonction uniquement, parfois sur toute la fonction. On parle alors de minimum/maximum **relatifs** (pour un interval de la fonction) et **absolus** (sur tout l'interval de la fonction).

#### Exemples

##### Exemple 1

On considère pour cette exemple, la fonction carré : $f(x) = x^2$ (à chaque fois qu'on va passer un nombre à cette fonction, elle va renvoyer ce nombre au carré).

Pour bien comprendre, on peut faire le tableau de valeurs :

| $x$    | -2 | 0  | 2  |
|--------|----|----|----|
| $f(x)$ | 4  | 0  | 4  |

A partir de la, on pourrait noter sous la forme de fonction, differentes valeurs de la fonction :

- $f(-2) = 4$ : L'image de -2 est 4.
- $f(2) = 4$ : L'image de 2 est 4.
- Les antécédents de 4 sont 2 et -2.



```python
import matplotlib.pyplot as plt
import numpy as np

# Définition de la fonction
def lin(x):
    return x**2

# Génération des valeurs pour x et y
x = np.linspace(-5, 5, 100)
y = lin(x)

# Création du graphique
fig, ax = plt.subplots()

# Tracer la courbe
ax.plot(x, y)

# Centrer l'axe des ordonnées et des abscisses sur la courbe
ax.spines['left'].set_position('zero')
ax.spines['right'].set_color('none')
ax.yaxis.tick_left()

ax.spines['bottom'].set_position('zero')
ax.spines['top'].set_color('none')
ax.xaxis.tick_bottom()

# Afficher les points (a, f(a)) et (b, f(b)) pour a < b
a = -2
b = 2
fa = lin(a)
fb = lin(b)
ax.scatter(a, fa, color='r', label=f'({a}, {fa})')
ax.scatter(b, fb, color='g', label=f'({b}, {fb})')

# Ajouter une légende
ax.legend()

# Ajout des labels et titre
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Fonction carré (x²)')

# Tracer les lignes pointillées

ax.plot([a,a], [0,fa], color='r', linestyle='--')
ax.plot([b,b], [0,fb], color='g', linestyle='--')
ax.plot([a,0], [fa,fa], color='r', linestyle='--')
ax.plot([b,0], [fb,fb], color='g', linestyle='--')

# Affichage du graphique
plt.show()
```


##### Exemple 2

On considère cette fonction : $f(x) = x^2 - 3$.

On peut faire les calculs sous forme de fonction tout en remplissant le tableau de valeurs au fur et a mesure de nos résultats :

$f(-2) = (-2)^2 - 3 = 1$

$f(0) = 0^2 - 3 = -3$

$f(1) = 1^2 - 3 = -2$

$f(3) = 3^2 - 3 = 6$


| $x$    | -2 | 0  | 1  | 3  |

```python
import matplotlib.pyplot as plt
import numpy as np

# Définition de la fonction
def fonction(x):
    return x**2-3

# Génération des valeurs pour x et y
# Calcul et trace 100 points pour la fonction
# sur l'interval [-5; 5]
x = np.linspace(-5, 5, 100)
y = fonction(x)

# Création du graphique
fig, ax = plt.subplots()

# Centrer l'axe des ordonnées et des abscisses sur la courbe
ax.spines['left'].set_position('zero')
ax.spines['right'].set_color('none')
ax.yaxis.tick_left()

ax.spines['bottom'].set_position('zero')
ax.spines['top'].set_color('none')
ax.xaxis.tick_bottom()

# Tracer la courbe
ax.plot(x, y)

# Ajout des labels et titre
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Fonction (x² - 3)')

# Affiche les graduations avec un intervalle de 1 sur les axes x et y
ax.set_xticks(np.arange(-5, 6, 1))
ax.set_yticks(np.arange(-5, 26, 1))

# Affiche la grille

# Affichage du graphique
plt.show()
```

## Fonctions de référence

### Fonctions affines

**Définitions** : $a$ et $b$ étant deux nombres fixés (_constants_).

- Fonction affine : $f(x) = ax+b$
- Fonction linéaire : $f(x) = ax$ ($b$ est égal à $0$).
- Fonction constante : $f(x) = b$ ($a$ est égal à $0$).

**Propriété 1** :

1. Une fonction affine est représentée par une droite.
2. Une fonction linéaire est représentée par une droite passant par l'origine.
3. Une fonction constantes est représentée par une droite parallèle à l'axe des abscisses.

**Propriété 2** :

- Si $a > 0$, alors $f$ est croissante.
- Si $a < 0$, alors $f$ est décroissante.
- Si $a = 0$, alors $f$ est constante.

**Propriété 3** :

- Soit une fonction affine $f$ définie par $f(x) = ax + b$ représentée dans un repère par une droite $d$. Les coordonnées $(x;y)$ d'un point $M$ appartenant à $d$ vérifient $y = ax + b$.

**Propriété 4** :

- La droite $(d)$ représentant la fonction $f$ définie par $f(x) = ax + b$ a pour coefficient directeur $a$ et pour ordinnée à l'origine $b$.

Pour l'_ordonnée à l'origine_, il suffit de regarder sur l'axe des ordonnées ; la ou la droite coupe l'axe. Pour le _coefficient directeur_ (pente), il faut regarder pour un pas de 1 entre deux points sur l'axe des abcisses, de combien est leur différence sur l'axe des ordonnées.

**Propriété 5** : _Propriété des accroissements_

- Si $A(x_A ; y_A)$ et $B(x_B;y_B)$ sont deux points sur la droite $(d)$ représentant la fonction $f$ définie par $f(x) = ax + b$ alors : $a = \frac{y_B - y_A}{x_B - x_A}$.

```python
%matplotlib widget
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider

# Créer la figure et les axes
fig, ax = plt.subplots()
plt.subplots_adjust(left=0.25, bottom=0.25)
ax.set_xlim(-10, 10)
ax.set_ylim(-10, 10)

# Coefficients initiaux pour a et b
a_init = 1
b_init = 0

# Créer la fonction affine et le tracé initial
x = range(-10, 11)
y = [a_init * xi + b_init for xi in x]
line, = ax.plot(x, y)

# Créer les curseurs pour a et b
axcolor = 'lightgoldenrodyellow'
ax_a = plt.axes([0.25, 0.1, 0.65, 0.03], facecolor=axcolor)
ax_b = plt.axes([0.25, 0.15, 0.65, 0.03], facecolor=axcolor)
slider_a = Slider(ax_a, 'a', -100, 100, valinit=a_init)
slider_b = Slider(ax_b, 'b', -10, 10, valinit=b_init)

# Fonction pour mettre à jour le tracé
def update(val):
    a = slider_a.val
    b = slider_b.val
    y = [a * xi + b for xi in x]
    line.set_ydata(y)
    fig.canvas.draw_idle()
    fig.suptitle(f'Fonctions affines (F(x) = {int(a)}x + {int(b)})', fontsize=14)

# Régler les bords des axes pour se croiser en zéro
ax.spines['left'].set_position(('data', 0))
ax.spines['bottom'].set_position(('data', 0))
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

# Lier les curseurs à la fonction de mise à jour
slider_a.on_changed(update)
slider_b.on_changed(update)

# Met un titre au graph
fig.suptitle(f'Fonctions affines (F(x) = {int(a_init)}x + {int(b_init)})', fontsize=14)

# Afficher la figure
plt.show()

```

### Fonction paire

Dans un repère orthogonal, la courbe représentative d'une fonction paire est symétrique par rapport à l'axe des ordonnées.

- Une fonction $f$ est **paire** lorsque $f(-x) = f(x)$.

**Exemple** : _La fonction carré, la fonction sinus_

$f(x) = x^2$

Cette fonction est : 

- décroissante sur l'intervalle $]-{\infty} ; 0]$,
- croissante sur l'intervalle $[0 ; +{\infty}[$.

$f(x) = sin(x)$

```python
import matplotlib.pyplot as plt
import numpy as np
import math

# Génération des valeurs pour x et y
# Calcul et trace 100 points pour la fonction
# sur l'interval [-5; 5]
x = np.linspace(-5, 5, 100)
y = np.cos(x)

# Création du graphique
fig, ax = plt.subplots()

# Centrer l'axe des ordonnées et des abscisses sur la courbe
ax.spines['left'].set_position('zero')
ax.spines['right'].set_color('none')
ax.yaxis.tick_left()

ax.spines['bottom'].set_position('zero')
ax.spines['top'].set_color('none')
ax.xaxis.tick_bottom()

# Tracer la courbe
ax.plot(x, y)

# Ajout des labels et titre
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Fonction sinus (sin(x))')

# Affiche les graduations avec un intervalle de 1 sur les axes x et y
ax.set_xticks(np.arange(-5, 5, 1))
ax.set_yticks(np.arange(-5, 5, 1))

# Affiche la grille

# Affichage du graphique
plt.show()
```

### Fonction impaire

Dans un repère orthogonal, la courbe représentative d'une fonction impaire est symétrique par rapport à l'origine.

- Une fonction $f$ est **impaire** lorsque $f(-x) = -f(x)$.

**Exemple** : _La fonction tangente_

$f(x) = tan(x)$

```python
import matplotlib.pyplot as plt
import numpy as np
import math

# Génération des valeurs pour x et y
# Calcul et trace 100 points pour la fonction
# sur l'interval [-5; 5]
x = np.linspace(-5, 5, 100)
y = np.tan(x)

# Création du graphique
fig, ax = plt.subplots()

# Centrer l'axe des ordonnées et des abscisses sur la courbe
ax.spines['left'].set_position('zero')
ax.spines['right'].set_color('none')
ax.yaxis.tick_left()

ax.spines['bottom'].set_position('zero')
ax.spines['top'].set_color('none')
ax.xaxis.tick_bottom()

# Tracer la courbe
ax.plot(x, y)

# Ajout des labels et titre
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Fonction tangente (tan(x))')

# Affiche les graduations avec un intervalle de 1 sur les axes x et y
ax.set_xticks(np.arange(-6, 6, 2))
ax.set_yticks(np.arange(-300, 300, 30))

# Affiche la grille

# Affichage du graphique
plt.show()
```

**Exemple** : _La fonction inverse_

$f(x) = \frac{1}{x}$

Cette fonction est : 

- décroissante sur l'intervalle $]-{\infty} ; 0[$,
- croissante sur l'intervalle $]0 ; +{\infty}[$.
- Elle a la particularité de ne pas être défini pour $x = 0$. Car on ne peut pas faire $1/0$.
