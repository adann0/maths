# Fonctions du second degré

Une fonction polynôme de degré 2 est définie sur $\mathbb{R}$ par :

\begin{equation}
f(x) = ax^2 + bx + c
\end{equation}

Par exemple, les fonctions suivantes sont des fonctions du second degré :

- $f(x) = 3x^2 + 7x + 3$
- $g(x) = \frac{1}{2}x^2 - 5x + \frac{3}{5}$
- $h(x) = 4 - 2x^2$
- $k(x) = (x - 4)(5 - 2x)$

Une fonction du second degré peut également s'écrire sous sa **forme canonique** :

\begin{equation}
f(x) = a(x - α)^2 + β
\end{equation}

ou α et β sont deux nombres réels.

**Propriété** :

- Si _a_ > _0_, _f_ admet un minimum pour _x_ = _α_. Ce minimum est égal à _β_.
- Si _a_ < _0_, _f_ admet un maximum pour _x_ = _α_. Ce maximum est égal à _β_.



```python
%matplotlib widget
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider

# Créer la figure et les axes
fig, ax = plt.subplots(figsize=(8,8))
plt.subplots_adjust(left=0.3, bottom=0.4, top=0.9, right=0.9)
ax.set_xlim(-10, 10)
ax.set_ylim(-50, 50)

# Coefficients initiaux pour a, b et c
a_init = 0
b_init = 0
c_init = 0

# Créer la fonction polynomiale et le tracé initial
x = np.linspace(-10, 10, 1000)
y = a_init * x**2 + b_init * x + c_init
line, = ax.plot(x, y)

# Créer les curseurs pour a, b et c
axcolor = 'lightgoldenrodyellow'
ax_a = plt.axes([0.25, 0.3, 0.65, 0.03], facecolor=axcolor)
ax_b = plt.axes([0.25, 0.25, 0.65, 0.03], facecolor=axcolor)
ax_c = plt.axes([0.25, 0.2, 0.65, 0.03], facecolor=axcolor)
slider_a = Slider(ax_a, 'a', -10, 10, valinit=a_init)
slider_b = Slider(ax_b, 'b', -10, 10, valinit=b_init)
slider_c = Slider(ax_c, 'c', -10, 10, valinit=c_init)

# Fonction pour mettre à jour le tracé
def update(val):
    a = slider_a.val
    b = slider_b.val
    c = slider_c.val
    y = a * x**2 + b * x + c
    line.set_ydata(y)
    fig.canvas.draw_idle()
    fig.suptitle(f'Polynomial Function (y = {a:.2f}x^2 + {b:.2f}x + {c:.2f})', fontsize=14)
    info_text = f'a = {a:.2f}, b = {b:.2f}, c = {c:.2f}'
    info_label.set_text(info_text)

# Régler les bords des axes pour se croiser en zéro
ax.spines['left'].set_position(('data', 0))
ax.spines['bottom'].set_position(('data', 0))
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

# Lier les curseurs à la fonction de mise à jour
slider_a.on_changed(update)
slider_b.on_changed(update)
slider_c.on_changed(update)

# Met un titre au graph
fig.suptitle(f'Fonction Polynome (f(x) = {a_init:.2f}x² + {b_init:.2f}x + {c_init:.2f})', fontsize=14)

# Afficher les valeurs de a, b et c
info_text = f'a = {a_init:.2f}, b = {b_init:.2f}, c = {c_init:.2f}'
info_label = ax.text(-10, -60, info_text, fontsize=12)

# Afficher la figure
plt.show()
```
