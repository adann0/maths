# Intégration

## Intégrale et aire

On sait que l'on peut facilement calculer à partir des formules des différents polygones leur air. Cependant pour calculer l'air d'une forme géométrie contenant de multiples courbes, c'est un peu plus complexe. C'est la que les **intégrales** vont nous être utiles.

Le symbole mathématique représentant l'intégration, le « S long » : ${\displaystyle \textstyle \int }$, est appelé signe d'intégration, signe intégral ou intégrateur.

**Définition** :

Soit $f$ une fonction continue et positive sur un intervalle $[a ; b]$.

On appelle **intégrale de $f$** sur $[a ; b]$ l'aire, exprimée en unité d'air (u.a.)., de la surface délimitée par la courbe représentative de la fonction $f$, l'axe des abscisses et les droites d'équations $x = a$ et $x = b$ ; et est noté :

\begin{equation}
\displaystyle \int_{a}^{b} f(x) \, \mathrm{d}x
\end{equation}

```python
import numpy as np
import matplotlib.pyplot as plt

# Définir la fonction à tracer
def f(x):
    return np.sin(x)

# Définir les bornes de l'intégrale
a = 1
b = 2

# Créer un tableau de valeurs de x
x = np.linspace(0, np.pi, 100)

# Tracer la fonction
plt.plot(x, f(x), 'b-', linewidth=2)

# Remplir l'aire sous la courbe
x_fill = np.linspace(a, b, 100)
y_fill = f(x_fill)
plt.fill_between(x_fill, y_fill, 0, alpha=0.2)

# Ajouter une étiquette pour l'aire sous la courbe
plt.text((a+b)/2, 0.3, 'Aire', ha='center', va='center', color='blue', fontsize=20)

# Ajouter des marqueurs pour les bornes a et b
plt.axvline(a, color='red', linestyle='--')
plt.axvline(b, color='red', linestyle='--')

# Définir les limites de l'axe des x et y
plt.xlim(0, np.pi)
plt.ylim(0, 1)

# Ajouter des étiquettes pour les bornes a et b
plt.text(a, -0.1, '$a$', ha='center', va='top', color='red')
plt.text(b, -0.1, '$b$', ha='center', va='top', color='red')
plt.text(a-0.2, 0.96, '$x = a$', ha='center', va='top', color='red')
plt.text(b+0.2, 0.96, '$x = b$', ha='center', va='top', color='red')


# Afficher le graphique
plt.show()
```

## Unité d'air

On a pas toujours une graduation définie. Parfois c'est une graduation d'unité "quelconque" définie comme ci-dessus. Pour savoir ce qu'est unité d'air ; il suffit de regarder quelle est l'unité sur l'axe des absicces, une unité sur l'axe des ordonnées ; et on obtient une unité d'air.

```python
import matplotlib.pyplot as plt

# Créer les coordonnées des coins de la case
x = [0, 0, 1, 1]
y = [0, 1, 1, 0]

# Créer la figure et les axes
fig, ax = plt.subplots()

# Tracer un rectangle
ax.add_patch(plt.Rectangle((0, 0), 1, 1, facecolor='green', alpha=0.5))

# Tracer une grille pour la visualisation
ax.grid(True)

# Ajouter un titre
ax.set_title('Mise en évidence de la case d\'une unité d\'aire')

# Ajouter des étiquettes d'axes
ax.set_xlabel('Axe X')
ax.set_ylabel('Axe Y')

# Modifier les limites des axes
ax.set_xlim(0, 10)
ax.set_ylim(0, 10)

# Afficher le graphique
plt.show()
```

## Intégrale et rectangles

```python
%matplotlib widget
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

# Définition de la fonction f(x) à intégrer
def f(x):
    return np.sin(x)

# Définition de la limite inférieure et supérieure de l'intégrale
a, b = 0, np.pi

# Définition du nombre initial de rectangles utilisés pour l'approximation de l'intégrale
N = 3

# Initialisation de la figure et des axes
fig, ax = plt.subplots()

# Tracé de la fonction f(x)
x = np.linspace(a, b, 1000)
ax.plot(x, f(x), 'r-', lw=2)

# Définition de la fonction d'animation
def animate(i):
    ax.cla()  # Efface le contenu des axes
    ax.plot(x, f(x), 'r-', lw=2)  # Trace la fonction f(x)
    
    # Calcul du nombre de rectangles pour l'étape courante
    n_rect = N * 2 ** i
    
    # Calcul de la largeur de chaque rectangle
    dx = (b - a) / n_rect
    
    # Définition des coordonnées x et y des sommets de chaque rectangle
    x_left = np.linspace(a, b - dx, n_rect)
    x_right = np.linspace(dx, b, n_rect)
    y = f(x_left)
    
    # Trace les rectangles
    ax.bar(x_left, y, width=dx, align='edge', alpha=0.5)
    
    ax.set_xlim([a, b])  # Définition des limites de l'axe des x
    ax.set_ylim([0, 1])  # Définition des limites de l'axe des y
    ax.set_title('Approximation de l\'intégrale de f(x) entre a = 0 et b = 3.3 (N = {})'.format(n_rect))

# Calcul du nombre d'étapes de l'animation (nombre de fois où le nombre de rectangles est doublé)
n_steps = int(np.log2(1000/N))

# Création de l'animation
anim = animation.FuncAnimation(fig, animate, frames=n_steps+1, interval=1000, repeat=True, blit=False)

# Affichage de l'animation
plt.show()
```

## Primitives

**Théorème** : Soit $f$ une fonction continue et positive sur un intervalle $[a;b]$. La fonction $F$ définie sur $[a;b]$ par $F(x)=\int_{a}^{x} f(t)dt$ est dérivable sur $[a;b]$ et sa dérivée est la fonction $f$.

\begin{equation}
F'(x) = f(x)
\end{equation}

**Définition** : $f$ est une fonction continue sur un intervalle I. On appelle **primitive** de $f$ sur I, une fonction $F$ dérivable sur I telle que $F' = f$.

<center><b>Primitives élémentaires</b></center>

| $f(x)$ | $D_f$ | $F(x)$ |
| :---: | :---: | :---: |
| $a$ | $\mathbb{R}$ | $ax$ | 
| $\frac{1}{x}$ | $\mathbb{R}^*$ | $\ln{|x|}$ |
| $x^a$ (avec $a \neq -1$) | $\begin{cases} \mathbb{R} &\mbox{si } a \in \mathbb{N} \\ \mathbb{R}^* &\mbox{si } a \in \mathbb{Z} \setminus \mathbb{N} \\ \mathbb{R}_+^* &\mbox{sinon} \end{cases}$ | $\frac{x^{a+1}}{a+1}$ |
| $a^x$ (avec $a>0$ et $a\neq1$) | $\mathbb{R}$ | $\frac{a^x}{\ln{a}}$ |
| $\ln{(x)}$ | $\mathbb{R}_+^*$ | $x(\ln{(x)}-1)$ |

<br>
<center><b>Primitives avec fonctions trigonométriques</b></center>

| $f(x)$ | $D$ | $D_f$ | $F(x)$ |
|:---------------------------------------:|:----:|:-:|:---:|
|            $\cos(x)$                     |   $R$  | $\mathbb {R}$ | $\sin(x)$ |
|            $\sin(x)$                     |   $R$  | $\mathbb {R}$ | $-cos(x)$ |
|    $\frac{1}{cos^2 x}$                   |  $$R \backslash \{\frac{\pi}{2} + k \pi \mid k\in \mathbb{Z}\}$$ | $\mathbb{R}$ | $\tan(x)$ |
|  $-\frac{1}{sin^2 x}$                    |  $R\backslash\{k\pi\mid k\in \mathbb{Z}\}$ | $\mathbb{R}$ | $\cot(x)$ |
|            $\tan(x)$                     |  $R\backslash\{\frac{\pi}{2}+k\pi \mid k\in \mathbb{Z}\}$ | $\mathbb{R}$ | $-\ln|\cos(x)|$ |
|            $\cot(x)$                     |  $R\backslash\{k\pi\mid k\in \mathbb{Z}\}$ | $\mathbb{R}$ | $\ln|\sin(x)|$ |
|     $\frac{1}{\sqrt{1-x^2}}$             |  $]-1,1[$ | $\mathbb{R}$ | $\arcsin(x)$ |
|    $\frac{1}{1+x^2}$                     |   $R$  | $\mathbb{R}$ | $\arctan(x)$ |
