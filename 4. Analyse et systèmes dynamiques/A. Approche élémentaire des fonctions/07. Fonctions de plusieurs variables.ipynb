# Fonction de plusieurs variables

En mathématiques et plus spécialement en analyse vectorielle, une fonction numérique à plusieurs variables réelles est une fonction dont l'ensemble de départ $E$ est une partie de $\mathbb{R}^{n}$. L'ensemble d'arrivée $F$ peut être $\mathbb{R}$ ou $\mathbb{R}^{p}$. Le second cas peut se ramener au premier cas en considérant qu'il s'agit en réalité de $p$ fonctions de $\mathbb{R}^{n}$ dans $\mathbb{R}$ appelées fonctions coordonnées.

La fonction est donc une relation associant à chaque n-uplet $x = (x_1, x_2, ..., x_n)$ élément de l'ensemble de départ un et un seul élément de l'ensemble d'arrivée, que l'on appelle image de $x$ par $f$ et que l'on note $f(x)$ ou $f(x_1, ..., x_n)$ :

\begin{equation}
{\displaystyle f:{\begin{array}{ccc}E&\longrightarrow &F\\(x_{1},\ldots ,x_{n})&\longmapsto &f(x_{1},\ldots ,x_{n})\end{array}}}
\end{equation}

En fonction d'une ou plusieurs variables de départ et/ou d'arrivée, on mettra un exposant correspondant au nombre de varriables prises et/ou retournées par la fonction :

\begin{equation}
{\displaystyle f:{\begin{array}{ccc}\mathbb{R}^n&\longrightarrow &\mathbb{R}\\(x_{1},\ldots ,x_{n})&\longmapsto &f(x_{1},\ldots ,x_{n})\end{array}}}
\end{equation}



Si l'on munit les deux espaces vectoriels $\mathbb{R}^{n}$ et $\mathbb{R}^{p}$ d'une norme, on peut étudier la continuité et la différentiabilité de telles fonctions. En fixant les $n$ variables réelles $(x_1, x_2, ..., x_n)$ sauf une, on se ramène à l'étude de fonctions d'une variable réelle, à valeurs dans $\mathbb {R}^{p}$ (ou même dans $\mathbb{R}$ , en considérant les $p$ fonctions coordonnées). Leurs dérivées, lorsqu'elles existent, s'appellent les dérivées partielles de la fonction de départ.

## Exemple

- $f(x,y) = x^2 - y^3 + x^2y$
- $f(x,y) = cos(x,y)$
- $f(x,y) = \sqrt{x^2+y^2}$

## Graphe (Plan)

Pour une fonction de 2 varriables, retournant 1 résultat ; on peut représenter les résultats de cette fonction sur un Plan comme ci-dessous.



${\displaystyle f:{\begin{array}{ccc}\mathbb{R}^2&\longrightarrow &\mathbb{R}\\(x,y)&\longmapsto &f(x,y) = \sqrt{x^2+y^2} = z\end{array}}}$

A chaque point $(x, y)$ de $\mathbb{R}^2$ (en largeur et en longueur) correspond un point $z$ sur $\mathbb{R}$ (en hauteur) représentant le résultat de la fonction.




```python
%matplotlib widget
import matplotlib.pyplot as plt
import numpy as np

# Créer une grille de données pour x et y
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)

# Calculer les valeurs de z
Z = np.sqrt(X**2 + Y**2)

# Créer une figure et un axe 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Tracer la surface en faisant varier les couleurs en fonction de z avec une colormap 'hsv'
surf = ax.plot_surface(X, Y, Z, cmap='hsv')

# Ajouter une étiquette d'axe
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

# Ajouter une barre de couleur
fig.colorbar(surf, shrink=0.5, aspect=5)

# Ajouter un titre
plt.title('Fonction f(x,y) = √(x² + y²)')

# Afficher le graphe
plt.show()
```

```python
%matplotlib widget
import matplotlib.pyplot as plt
import numpy as np

# Définir la fonction f(x,y)
def f(x, y):
    return x * np.exp(-x**2 - y**2)

# Définir les données pour x et y
x = np.linspace(-2, 2, 50)
y = np.linspace(-2, 2, 50)

# Calculer les valeurs de f(x,y) pour chaque point de la grille
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# Créer une figure et un axe 3D
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111, projection='3d')

# Tracer la surface
surf = ax.plot_surface(X, Y, Z, cmap='cividis', alpha=0.7)

# Tracer les tranches pour f(0,y) en rouge
ax.plot(np.zeros_like(y), y, f(0,y), 'r', linewidth=4)

# Tracer les tranches pour f(x,0) en bleu
ax.plot(x, np.zeros_like(x), f(x,0), 'b', linewidth=2)

# Ajouter des étiquettes d'axe et un titre
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x,y)')
plt.title('Plan de la fonction f(x,y) avec des tranches pour f(0,y) et f(x,0)')

# Ajouter une barre de couleur
fig.colorbar(surf, shrink=0.5, aspect=5)

# Afficher le graphe
plt.show()
```

```python
%matplotlib widget
import matplotlib.pyplot as plt
import numpy as np

# Définition de la fonction
def f(x, y):
    return x**2 + y**2

# Création des données d'entrée
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# Calcul des valeurs de sortie de la fonction
Z = f(X, Y)

# Tracé des lignes de niveau
levels = np.linspace(0, 25, 11)  # Niveaux de 0 à 25, espacés de 2.5
plt.contour(X, Y, Z, levels=levels, colors='black')

# Tracé des courbes de niveau
plt.contourf(X, Y, Z, levels=levels, cmap='cool')

# Paramètres du graphique
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Lignes et courbes de niveau')
plt.colorbar()
plt.show()
```

```python
%matplotlib widget
import matplotlib.pyplot as plt
import numpy as np

# Définition de la fonction
def f(x, y):
    return x**2 + y**2

# Création des données d'entrée
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# Calcul des valeurs de sortie de la fonction
Z = f(X, Y)

# Tracé du plan en 3D
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.plot_surface(X, Y, Z, cmap='coolwarm')

# Tracé des courbes de niveau sur le plan
levels = np.linspace(0, 25, 11)  # Niveaux de 0 à 25, espacés de 2.5
cf = ax.contourf(X, Y, Z, levels=levels, cmap='cool', alpha=0.5)

# Tracé des lignes de niveau sur le plan
ax.contour(X, Y, Z, levels=levels, colors='black')

# Paramètres du graphique
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.title('Lignes et courbes de niveau')
plt.colorbar(cf)
plt.show()

```
