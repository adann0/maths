# Loi binomiale

## Epreuve de Bernoulli

<span style="color: #ff4d4d;">Une **épreuve de Bernoulli** est une expérience aléatoire à deux issues que l'on peut nommer "succès" ou "échec".</span>

**Exemple** :

- _Le jeu du file ou face : on considère par exemple comme succès "obtenir pile" et comme echec "obtenir face"._
- _On lance un dé : on considère par exemple comme succès "obtenir un six" et comme échec "ne pas obtenir un six"._

## Loi de Bernoulli

<span style="color: #ff4d4d;">Une **loi de Bernoulli** est une loi de probabilité qui suit le schéma suivant :</span>

- <span style="color: #ff4d4d;">la probabilité d'obtenir un succès est égale à $p$,</span>
- <span style="color: #ff4d4d;">la probabilité d'obtenir un échec est égale à $1-p$. $p$ est appelé le _paramètre_ de la loi de Bernouilli.</span>

**Exemple** :

- _Le jeu du file ou face : $p = \frac{1}{2}$_
- _On lance un dé : $p = \frac{1}{6}$_

Habituellement, on rescence les résultats de la loi de probabilité dans un tableau :

| $$x_i$$        | 1     | 0     |
|:--------------:|:-----:|:-------:|
| $$P(X=x_i)$$   | $$p$$ | $$1-p$$ |

_On écrit pas X=succès ou X=echec, mais X=1 ou X=0._

**Exemple** :

_Si on souhaite compléter notre tableau de probabilité avec l'exemple du dé, ça nous donnerait :_

| $$x_i$$        | 1     | 0     |
|:--------------:|:-----:|:-------:|
| $$P(X=x_i)$$   | $$\frac{1}{6}$$ | $$\frac{5}{6}$$ |


<center><div style="margin-left: 28px; border: 1px solid white; padding: 14px; display: inline-block;">
    <p><span style="color: #ff4d4d;">La variable aléatoire $X$ suit la loi de Bernouilli de paramètre $p$, alors :</span></p>
    <p style="text-align: center; font-size: 1.2em; margin: 0;">
        $E(X) = p$ <br>
        $V(X) = p(1-p)$ <br>
        $σ(X) = \sqrt{p(1-p)}$
    </p>
</div></center>




## Répétition d'expériences

<span style="color: #ff4d4d;">Plusieurs expériences sont **identiques et indépendantes** si :</span>

- <span style="color: #ff4d4d;">elles ont les mêmes issues,</span>
- <span style="color: #ff4d4d;">les probabilités de chacune des issues ne changent pas d'une expérience à l'autre.</span>

**Exemple** :

_Les lancés d'une pièce de monnaie sont identiques et independantes car une experience d'influence pas sur l'autre. Si on prend une urne avec des boules, et que l'on ne remet pas les boules entre chaques lancés ; les experiences ne seront ni independentes, ni identiques ; car les probabilités de tirer chaque boules seront modifiées entre les deux tirages._

## Schéma de Bernoulli

<span style="color: #ff4d4d;">Un **schéma de Bernoulli** est la répétition de $n$ épreuves de Bernoulli identiques et indépendantes pour lesquelles la probabilité du succès est $p$.</span>

_Faire bien la différence entre une Epreuve de Bernouilli et un Schema de Bernouilli. Un Schema de Bernouilli est bien la répétition de n epreuves de Bernouilli identiques et indépendantes (voir un arbre)_.

**Exemple** :

_n = 10 lancés d'une pièce de monaie et la probabilité de succès est p = 1/2_ On va donc pouvoir établir une loi binomiale qui va nous permettre de faire des calculs de probabilités, par exemple la chance de faire que des piles, ou de faire que des faces.

## Loi binomiale

<span style="color: #ff4d4d;">On réalise un schéma de Bernoulli composé de _n_ épreuves de Bernoulli identiques et indépendantes.</span>

<span style="color: #ff4d4d;">Une **loi binomiale** est une loi de probabilité définie sur l'ensemble {0 ; 1 ; 2 ; ... ; n} qui donne le nombre de succès de l'expérience.</span>

<span style="color: #ff4d4d;">Un arbre contenant $n$ générations correspond à une loi binomiale $B(n, p)$.</span>

**Exemple** :

_En reprenant l'exemple précédent, la loi de probabilité est définie sur l'ensemble {0 ; 1 ; 2 ; ... ; 10} car on peut avoir entre 0 et 10 succès, et on dira que n et p sont les paramètres de la loi binomiale. On notera B(10, 1/2) la loi binomiale de paramètre n et p. n c'est le nombre de répétition de l'experience, p c'est la probabilité de succès de chaque épreuve aléatoire._

<span style="color: #ff4d4d;">On réalise une expérience suivant un schéma de Bernoulli de paramètre $n$ et $p$. On appelle **coefficient binomial** ou **combinaison de k parmi n**, le nombre de chemins conduisant à $k$ succès parmi $n$ épreuves sur l'arbre représentant l'expérience. Ce nombre se note : ${n \choose k}$.</span>

<span style="color: #ff4d4d;">La variable aléatoire X suit la loi binomiale $B(n;p)$ :</span>

\begin{equation}
{\displaystyle \mathbb {P} (X=k)={n \choose k}\,p^{k}(1-p)^{n-k}.}
\end{equation}

**Exemple** :

_Une boite contient 2 jetons jaunes et 3 jetons rouges._

_On tire au hasard 4 fois de suite un jeton en le remettant à chaque fois dans la boite._

_X est la variable aléatoire qui associe le nombre de jetons rouges obtenus._

_On a tiré 3 jetons rouges et 1 jetons jaune. On souhaite connaitre la probabilité que cette situation se produise (que l'on ai tiré en tout 3 jetons rouges et 1 jeton jaune)._

- _$n = 4$ (on a effectué 4 tirage identiques et indépendants),_
- _$p = \frac{3}{5}$ (la probabilité de succès, c'est à dire d'avoir un jeton rouge, est de 3/5),_
- _On associe a X une loi de Bernouilli $B(4 ; \frac{3}{5})$,_
- _On cherche P(X = 3), on cherche alors le nombre de combinaison pour obtenir 3 succès parmis 4 tirages (${4 \choose 3}$)_
- _Il nous faut maintenant pour 1 des chemin, calculer la probabilité d'avoir 3 succès. Il faut donc faire ($\frac{3}{5} \times \frac{3}{5} \times \frac{3}{5}$) soit ($(\frac{3}{5})^3$) que l'on retrouve dans la formule finale ci-dessous._
- _$(1 - p)$ revient a recuperer l'echec, donc $\frac{2}{5}$ dans notre cas._


- _On obtient donc : $\mathbb{P}(X=3)={4 \choose 3} \times (\frac{3}{5})^{3} \times (\frac{2}{5})$_
