# Variables aléatoires

## Variables aléatoires

<span style="color: #ff4d4d;">Une **variable aléatoire** X est une fonction définie sur un univers Ω et à valeur dans $\mathbb{R}$.</span>

**Exemple** :

_Soit l'expérience aléatoire : "On lance un dé à six faces et on regarde le résultat."_

_L'ensemble de toutes les issues possibles Ω = {1 ; 2 ; 3 ; 4 ; 5 ; 6}._

_On considère le jeu suivant :_

- _Si le résultat est pair, on gagne 2 euros._
- _Si le résultat est 1, on gagne 3 euros._
- _Si le résultat est 3 ou 5, on perd 4 euros._

_On va définir ainsi une variable X qui donne le gain à ce jeu : elle peut donc renvoyer comme valeur 2, 3 ou -4 correspondant au gain._

_X est appelé une **varriable aléatoire** qui prend ses valeurs dans Ω = {1 ; 2 ; 3 ; 4 ; 5 ; 6} de façon aléatoire._

- $X(1) = 3$
- $X(2) = 2$
- $X(3) = -4$
- $X(4) = 2$
- $X(5) = -4$
- $X(6) = 2$

_On voit ici qu'une variable aléatoire est bien une fonction._

## Loi de probabilité

<span style="color: #ff4d4d;">Soit une variable aléatoire X prenant les valeurs $x_1, x_2, \dots, x_n$. La **loi de probabilité** de X associe à toute valeur $x_i$ la probabilité $P(X = x_i) = p_i$.</span>

**Exemple** :

_On reprend le jeu présenté précédement._

- $P(X=2) = \frac{1}{2}$
- $P(X=3) = \frac{1}{6}$
- $P(X=-4) = \frac{2}{6}$

_En général, on met ces valeurs dans un tableau :_

| $$x_i$$    |  2  |  3  | -4 |
|:------:|:---:|:---:|:---|
|$$P(x=x_i)$$| $$\frac{3}{6}$$ | $$\frac{1}{6}$$ | $$\frac{2}{6}$$ |

## Espérance

<span style="color: #ff4d4d;">L'esperance mathématique de X est :

\begin{equation}
E(X) = p_1x_1 + p_2x_2 + \dots + p_nx_n
\end{equation}

**Exemple** :

_Dans jeu précédent, l'espérance peut nous permettre de calculer combien on peut "esperer" gagner ou perdre en jouant un très grand nombre de fois au jeu._ 

_C'est un peu une moyenne, mais il faut imaginer que l'on utilise pas la formule de la moyenne en probabilités : on utilise la formule de l'esperance pour calculer la valeur attendue d'une variable aléatoire lors d'un très grand nombre de tirages. Additionner les gains de chaque tirages proportionnellement a leur taux de parution reviendrait a utiliser le mauvais outil mathématique._

- $E(X) = \frac{1}{2} \times 2 + \frac{1}{6} \times 3 + \frac{1}{3} \times -4 = \frac{1}{6}$

_On peut donc esperer gagner 1/6 euros par parties._

## Variance et Ecart-Type

<span style="color: #ff4d4d;">La variance et l'écart type permettent de calculer la dispersion des valeurs autour de l'esperance.

<span style="color: #ff4d4d;">La **variance** de X est :

\begin{equation}
V(X) = p_1(x_1 - E(X))^2 + p_2(x_2 - E(x))^2 + \dots + p_n(x_n - E(X))^2
\end{equation}

<span style="color: #ff4d4d;">L'**équart-type** de X est :

\begin{equation}
σ(X) = \sqrt{V(X)}
\end{equation}

**Exemple** :

_Toujours avec le jeu précédent._

- $V(X) = \frac{1}{2}(2 - \frac{1}{6})^2 + \frac{1}{6}(3 - \frac{1}{6})^2 + \frac{1}{3}(-4 - \frac{1}{6})^2 \approx 8.8$
- $σ(X) = \sqrt{V(X)} = \sqrt{8.8} \approx 2.97$

_On voit que par rapport aux gains, l'équart type est très grands. Ce qui signifie que l'on a beaucoup de chances de perdre de l'argent._

## Changement de variable

\begin{equation}
E(aX+b) = aE(X)+b
\end{equation}
\begin{equation}
V(aX+b) = a^2V(X)
\end{equation}

**Exemple** :

_Prenons l'exemple suivant, les calculs seraient complexes :_

| $$x_i$$    |        8.98        |        8.99        |       9.05        |
|:----------:|:---------------:|:---------------:|:----------------|
|$$P(x=x_i)$$| 0.4 | 0.1 | 0.5 |

_On peut simplifier les calculs, on va utiliser une equation qui va nous permettre de ne plus avoir de decimaux ; ici elle est donné, la plupart du temps il faudra biensur trouver cette equation :_

- $Y = 100X - 898$

| $$y_i$$    |        0        |        1        |       7        |
|:----------:|:---------------:|:---------------:|:----------------|
|$$P(y=y_i)$$| 0.4 | 0.1 | 0.5 |

_On commence ensuite par calculer l'esperance de Y :_

- $E(Y) = 0.1 + 7 \times 0.5 = 3.6$

_L'esperance de la variable Y est égale a 3.6._

_On peut ensuite calculer l'esperance de X._

- $E(Y) = E(100X - 898)$
- $3.6 = 100E(X) - 898$
- $E(X) = 9.016$
